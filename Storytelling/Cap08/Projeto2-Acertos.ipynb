## Visualização 1

Problema: Gráfico de Barras com Escala Inapropriada.

Erro: Usar uma escala que não começa em zero, o que pode induzir a interpretações erradas da magnitude das diferenças. Neste exemplo, o eixo Y começa em 40 ao invés de 0, o que pode induzir a interpretações erradas sobre a magnitude das diferenças entre as categorias, dando a impressão de que as diferenças são mais significativas do que realmente são.

```python
# Dados de exemplo
categorias = ['Categoria A', 'Categoria B', 'Categoria C', 'Categoria D']
valores = [55, 49, 90, 76]

# Criando o gráfico de barras
plt.figure(figsize = (10, 6))
plt.bar(categorias, valores, color = 'skyblue')

# Definindo uma escala para o eixo Y
plt.ylim(40, 100) 

# Labels
plt.title('Gráfico de Barras')
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.grid(axis = 'y')
plt.show()
```

> Versão Corrigida:

```python
# Dados de exemplo
categorias = ['Categoria A', 'Categoria B', 'Categoria C', 'Categoria D']
valores = [55, 49, 90, 76]

# Criando o gráfico de barras corrigido
plt.figure(figsize=(10, 6))
plt.bar(categorias, valores, color='skyblue')

# Definindo uma escala apropriada para o eixo Y (começando em 0)
plt.ylim(0, 100)  # Iniciando o eixo Y em 0

plt.title('Gráfico de Barras com Escala Correta')
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.grid(axis='y')
plt.show()
```

## Visualização 2

Problema: Gráfico de Pizza com Muitas Categorias.

Erro: Utilizar um gráfico de pizza quando há muitas categorias, tornando difícil a comparação entre as partes. Isso pode confundir e sobrecarregar visualmente quem analisa o gráfico, dificultando a interpretação clara das proporções entre as categorias.

```python
# Dados de exemplo
categorias = ['Cat A', 'Cat B', 'Cat C', 'Cat D', 'Cat E', 'Cat F', 'Cat G', 'Cat H', 'Cat I', 'Cat J', 'Cat K', 'Cat L']
valores = [5, 7, 4, 6, 7, 8, 5, 9, 3, 2, 4, 5]

# Criando o gráfico de pizza
plt.figure(figsize = (10, 8))
plt.pie(valores, labels = categorias, autopct = '%1.1f%%', startangle = 140)

# Labels
plt.title('Gráfico de Pizza')
plt.show()
```

> Versão Corrigida:

```python
# Criando um gráfico de barras horizontal como alternativa ao gráfico de pizza para melhorar a visualização
plt.figure(figsize = (10, 8))
plt.barh(categorias, valores, color = 'lightblue')
plt.xlabel('Valores')
plt.title('Gráfico de Barras Horizontal - Melhor para Muitas Categorias')
plt.grid(axis='x')
plt.show()
```

## Visualização 3

Problema: Gráfico de Linhas para Dados Categóricos.

Erro: Utilizar um gráfico de linhas para representar dados categóricos não ordenados, o que sugere uma relação de ordem onde ela não existe. Este tipo de visualização sugere uma relação de ordem ou continuidade entre as categorias, onde, na realidade, não existe. Tal representação pode levar a interpretações equivocadas sobre a natureza dos dados, insinuando, por exemplo, que a transição de uma categoria para outra segue uma progressão lógica ou temporal, o que não é o caso para dados puramente categóricos. 

```python
# Dados categóricos 
categorias = ['Categoria A', 'Categoria B', 'Categoria C', 'Categoria D']
valores = [20, 35, 30, 45]

# Mapeando categorias para um valor numérico para poder plotar no gráfico de linhas
categorias_map = range(len(categorias))

# Cria o plot
plt.figure(figsize = (10, 6))
plt.plot(categorias_map, valores, marker = 'o', linestyle = '-', color = 'b')

# Ajustando os rótulos do eixo x para mostrar as categorias
plt.xticks(categorias_map, categorias)

# Labels
plt.title('Gráfico de Linhas')
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.grid(axis = 'y')
plt.show()
```

> Versão Corrigida:

```python
# Criando um gráfico de barras para representar melhor os dados categóricos
plt.figure(figsize = (10, 6))
plt.bar(categorias, valores, color = 'blue')
plt.title('Gráfico de Barras Para Dados Categóricos')
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.grid(axis = 'y')
plt.show()
```

## Visualização 4

Problema: Uso de Cores Não Intuitivas

Erro: Escolher uma paleta de cores que não oferece contraste suficiente ou que não é intuitiva para o significado dos dados. Utilizamos uma paleta de cores que não oferece contraste suficiente ou que não é intuitiva para representar os níveis de satisfação, de "Muito Satisfeito" a "Muito Insatisfeito". As cores escolhidas não seguem uma lógica intuitiva que possa indicar, de forma imediata, uma progressão ou regressão no nível de satisfação, como seria esperado em uma representação visual eficaz desse tipo de dado. 

```python
# Dados de exemplo
categorias = ['Muito Satisfeito', 'Satisfeito', 'Neutro', 'Insatisfeito', 'Muito Insatisfeito']
valores = [25, 35, 20, 15, 5]

# Definindo uma paleta de cores
cores = ['grey', 'orange', 'blue', 'yellow', 'green']

# Cria o plot
plt.figure(figsize = (10, 6))
plt.bar(categorias, valores, color = cores)

# Labels
plt.title('Nível de Satisfação (%)')
plt.xlabel('Nível de Satisfação')
plt.ylabel('Porcentagem (%)')
plt.xticks(rotation = 45)
plt.grid(axis = 'y')
plt.show()
```

> Versão Corrigida:

```python
# Corrigindo a paleta de cores para seguir uma ordem lógica de sentimentos
cores_corrigidas = ['green', 'lightgreen', 'grey', 'lightcoral', 'red']

# Criando o gráfico de barras com a nova paleta de cores
plt.figure(figsize = (10, 6))
plt.bar(categorias, valores, color = cores_corrigidas)

plt.title('Nível de Satisfação (%)')
plt.xlabel('Nível de Satisfação')
plt.ylabel('Porcentagem (%)')
plt.xticks(rotation = 45)
plt.grid(axis='y')
plt.show()
```

## Visualização 5

Problema: Sobrecarga de Informação.

Erro: Incluir muitos elementos visuais em uma única visualização, tornando-a confusa e difícil de interpretar. Este é um exemplo que ilustra a sobrecarga de informação em uma visualização, combinando múltiplas linhas de dados sobre vendas e custos em um único gráfico de linhas. Este gráfico contém uma quantidade excessiva de elementos visuais, como diferentes estilos de linhas e marcadores para representar várias séries de dados, tornando difícil a interpretação e comparação rápida entre as categorias de dados. Tal complexidade pode confundir o público, dificultando a identificação de tendências, padrões ou diferenças significativas. 

```python
# Dados de exemplo
data = {
    'Mes': ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho'],
    'Vendas A': [200, 150, 300, 250, 270, 300],
    'Vendas B': [220, 170, 310, 265, 290, 320],
    'Vendas C': [210, 160, 305, 255, 280, 310],
    'Custo A': [100, 80, 150, 130, 140, 150],
    'Custo B': [110, 90, 155, 135, 145, 155],
    'Custo C': [105, 85, 152, 132, 138, 152],
}

# Cria o dataframe
df = pd.DataFrame(data)

# Plot
plt.figure(figsize = (14, 8))

# Loop para a formatação
for column in df.columns[1:]:
    if 'Vendas' in column:
        plt.plot(df['Mes'], df[column], marker = 'o', linestyle = '-', label = column)
    else:
        plt.plot(df['Mes'], df[column], marker = 'x', linestyle = '--', label = column)

# Labels
plt.title('Vendas x Custo')
plt.xlabel('Mês')
plt.ylabel('Valor')
plt.legend()
plt.grid(True)
plt.show()
```

> Versão Corrigida:

```python
# Simplificando o gráfico dividindo em dois subplots para separar visualmente vendas de custos
fig, (ax1, ax2) = plt.subplots(2, 1, figsize = (14, 16))

# Plot de vendas
for column in df.columns[1:4]:  # Selecionando apenas colunas de vendas
    ax1.plot(df['Mes'], df[column], marker = 'o', linestyle = '-', label = column)

ax1.set_title('Vendas por Mês')
ax1.set_ylabel('Valor das Vendas')
ax1.legend()
ax1.grid(True)

# Plot de custos
for column in df.columns[4:]:  # Selecionando apenas colunas de custos
    ax2.plot(df['Mes'], df[column], marker = 'x', linestyle = '--', label = column)

ax2.set_title('Custos por Mês')
ax2.set_xlabel('Mês')
ax2.set_ylabel('Valor dos Custos')
ax2.legend()
ax2.grid(True)
plt.show()
```

## Visualização 6

Problema: Ignorar a Legenda ou Torná-la Confusa.

Erro: Não incluir legenda ou criar uma legenda que não corresponde adequadamente aos dados apresentados. Este é um exemplo de como uma legenda pode ser confusa e não corresponder adequadamente aos dados apresentados. Apesar de o gráfico mostrar duas séries de dados empilhadas representando "Grupo 1" e "Grupo 2", a legenda foi intencionalmente rotulada de maneira inadequada como "Dados X" e "Dados Y", o que pode causar confusão e dificultar o entendimento claro daquilo que cada cor no gráfico realmente representa. 

```python
# Dados de exemplo
categorias = ['Categoria A', 'Categoria B', 'Categoria C']
valores1 = [10, 20, 15]
valores2 = [25, 7, 18]

# Barras empilhadas
plt.figure(figsize = (10, 6))
plt.bar(categorias, valores1, color = 'blue', label = 'Grupo 1')
plt.bar(categorias, valores2, bottom = valores1, color = 'red', label = 'Grupo 2')

# Criando a legenda 
plt.legend(title = 'Legenda', labels = ['Dados X', 'Dados Y'])

# Labels
plt.title('Categorias x Valores')
plt.xlabel('\nCategorias')
plt.ylabel('Valores')
plt.show()
```

> Versão Corrigida:

```python
# Dados de exemplo
categorias = ['Categoria A', 'Categoria B', 'Categoria C']
valores1 = [10, 20, 15]
valores2 = [25, 7, 18]

# Corrigindo o gráfico com uma legenda apropriada
plt.figure(figsize = (10, 6))
plt.bar(categorias, valores1, color = 'blue', label = 'Grupo 1')
plt.bar(categorias, valores2, bottom = valores1, color = 'red', label = 'Grupo 2')

# Legenda correta
plt.legend(title = 'Legenda')

plt.title('Categorias x Valores')
plt.xlabel('\nCategorias')
plt.ylabel('Valores')
plt.show()
```

## Visualização 7

Problema: Comparação Injusta e Desproporcional.

Erro: Comparar grupos de dados que não estão em escalas comparáveis ou que deveriam ser normalizados para uma comparação justa. O exemplo ilustra uma comparação injusta ao representar vendas de uma loja A (em unidades) e visitantes de um site B (em milhares) no mesmo gráfico sem normalizar ou ajustar as escalas para refletir uma comparação justa e precisa. Esse arranjo sugere uma comparação direta entre dois conjuntos de dados que diferem significativamente em magnitude e natureza, o que pode levar a interpretações errôneas sobre o desempenho relativo ou a importância de um conjunto de dados em relação ao outro. 

```python
# Dados de exemplo
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril']
vendas = [950, 1050, 1100, 1150] 
visitantes = [4500, 4700, 4800, 5000]  

# Área de plotagem
fig, ax1 = plt.subplots(figsize = (10, 6))

# Gráfico de barras para vendas
ax1.bar(meses, vendas, color = 'g', label = 'Vendas (Loja A)')
ax1.set_xlabel('Mês')
ax1.set_ylabel('Vendas (Unidades)', color = 'g')

# Criando um segundo eixo y para os visitantes
ax2 = ax1.twinx()
ax2.plot(meses, visitantes, color = 'b', marker = 'o', linestyle = '-', label = 'Visitantes (Site B)')
ax2.set_ylabel('Visitantes (Milhares)', color = 'b')

# Título e legenda
plt.title('Vendas vs. Visitantes')
ax1.legend(loc = 'upper left')
ax2.legend(loc = 'upper right')
plt.show()
```

> Versão Corrigida:

```python
# Aplicando a transformação de log nas duas variáveis
vendas_log = np.log(vendas)
visitantes_log = np.log(visitantes)

# Área de plotagem com transformação de log
fig, ax1 = plt.subplots(figsize = (10, 6))

# Gráfico de barras para vendas com log
ax1.bar(meses, vendas_log, color = 'g', label = 'Vendas (Loja A) - Log')
ax1.set_xlabel('Mês')
ax1.set_ylabel('Log de Vendas (Unidades)', color = 'g')

# Criando um segundo eixo y para os visitantes com log
ax2 = ax1.twinx()
ax2.plot(meses, visitantes_log, color = 'b', marker = 'o', linestyle = '-', label = 'Visitantes (Site B) - Log')
ax2.set_ylabel('Log de Visitantes (Milhares)', color = 'b')

# Título e legenda
plt.title('Vendas vs. Visitantes (Log)')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.show()
```
